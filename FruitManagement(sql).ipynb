```python
# Fruit Management System

import mysql.connector

mydb=mysql.connector.connect(host='localhost',user='root',password='',database='college')
mycursor=mydb.cursor()

while True:
    print("-------------------------------------------------------------------------------------------")
    print("1.Add/Sell fruit ")
    print("2.View all Fruits ")
    print("3.Update quantity of fruit")
    print("4.Search for a fruit")
    print("5.Buy a fruit")
    print("6.delete a fruit")
    print("7.Exit")
    
    ch=int(input("Enter Your Choice:"))
    
    if ch==1:
            a=input("Enter Fruit name:")
            b=int(input("Enter Quantity:"))
            c=int(input("Enter Price per kg:"))
        
            sql="insert into fruitm values(%s ,%s, %s)"
            val=(a,c,b)
        
            mycursor.execute(sql,val)
            mydb.commit()
            print(a,"Added Successfully")
            
    elif ch==2:
        sql='Select * from fruitm;'
        mycursor.execute(sql,) 
        myresult=mycursor.fetchall()
        for i in myresult:
            print(i)
        mydb.commit()   
        print("Successfully Viewed")
            
    elif ch==3:
            a=input("Enter Fruit name:")
            c=int(input("Enter New Quantity :"))
            
            
            sql="select fquantity from fruitm where fruitname =%s;"
            val=(a,)
            mycursor.execute(sql,val)

             
            myresult=mycursor.fetchall()
            
            for i in myresult:
                old=int(i[0])
            total=old+int(c)
            sql="update fruitm set fquantity=%s where fruitname =%s;"
            val=(total,a)
            mycursor.execute(sql,val)
            mydb.commit()
    
    
    elif ch==4:
        n=input("Enter Fruit name :")
        
        sql="select * from fruitm where fruitname=%s"
        val=(n,)
        mycursor.execute(sql,val)
        myresult=mycursor.fetchall()
        if not myresult:
            print(f'{n} is Not Found!')
        else:
            print(f'{n} is Found.')
        mydb.commit()
        
        
        
    elif ch==5:
        ff=input("Enter fruit name :")
        q=input("Enter Quantity :")
        sql="select fquantity from fruitm where fruitname=%s"
        val=(ff,)
        mycursor.execute(sql,val)
        mydb.commit()
            
    elif ch==6:
        ff=input("Enter fruit name :")
        sql="delete from fruitm where fruitname=%s;"
        val=(ff,)
        mycursor.execute(sql,val)
        
        mydb.commit()
        print(ff,"deleted succesfully")
        
        
    elif ch==7:
        print("Visit Again...")
        break
            
    else :
        print("Wrong Choice!!!!!")
```
